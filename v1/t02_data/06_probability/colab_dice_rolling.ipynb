# Dice Rolling

## Python code to generate data from simulated dice rolling

---



We first import the necessary packages for our dice rolling simulation

```
import random
import numpy as np
import matplotlib.pyplot as plt
```

### We can use Python's random library to "roll" a die, randomly choosing an integer from 1 to 6

```
def roll6():
  return random.randint(1, 6)
```

If we repeatedly call this function, we can simulate rolling a dice once

```
roll6()
```



---

### We can also roll more dice, more times

By uniformly selecting random numbers from different ranges, we can simulate rolling dice with different numbers of sides. Using a Python dictionary, we can also store the results of rolling such dice many times, keeping track of the distribution of values.

```
def rollDice(numRolls = 1, numSides = 6):
  """ Rolls a dice with numSides number of sides (labeled from 1 to numSides)
      a total of numRolls times. 
      
      Returns a dictionary tracking the number of times each value was rolled.
  """
  
  # a dictionary for the number of rolls of each value on the die
  diceRolls = {}
  for side in range(1, numSides+1):
      diceRolls[side] = 0
  
  # roll the die numRolls times
  for roll in range (numRolls):
    roll = random.randint(1, numSides)
    diceRolls[roll] += 1
  return diceRolls
```

For instance, we can roll a 6-sided die 100 times:

```
rollDice(100, 6)
```

Or a 20-sided die 10000 times:

```
rollDice(10000, 20)
```



---

### We can also graph the results to better visualize the distribution of rolls

For instance, when we roll a 6-sided die 100 times, we might find:

```
diceToGraph = rollDice(100,6)
print(diceToGraph)

N = len(diceToGraph)

ind = np.arange(N)
barHeights = list(diceToGraph.values())
xLabels = list(diceToGraph.keys())
yAxisHeight = max(barHeights)
width = 0.5

plt.bar(ind, barHeights, align='center')

plt.ylabel('Number of Rolls')
plt.title('Dice Rolling')
plt.xticks(ind, xLabels)
plt.yticks(np.arange(0, yAxisHeight, 5))
plt.show()
```

In general, we can define a function to let us graph these results as follows:

```
def rollsToGraph(numRolls = 10000, numSides = 20):
  diceToGraph = rollDice(numRolls, numSides)
  print(diceToGraph)

  N = len(diceToGraph)

  ind = np.arange(N)
  barHeights = list(diceToGraph.values())
  xLabels = list(diceToGraph.keys())
  yAxisHeight = max(barHeights)
  width = 0.5

  plt.bar(ind, barHeights, align='center')

  plt.ylabel('Number of Rolls')
  plt.title('Dice Rolling')
  plt.xticks(ind, xLabels)
  plt.yticks(np.arange(0, yAxisHeight, yAxisHeight//10))
  plt.show()
```

```
rollsToGraph()
```
