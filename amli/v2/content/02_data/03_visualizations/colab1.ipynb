##Python Libraries

**Numpy** is a python library made for scientific computing and array manipulation. We won't look at it too much in this unit, but we will be using some of its basic functionality. 

**Matplotlib** is a python library made for basic graphing. We use it in this unit to create most of the visualizations we will be using.

**Seaborn** is another python library for graphing. It does heatmaps very well, whereas Matplotlib requires much more complicated code to get a heatmap working. Thus, we will be using it for heatmaps.

Run the code block below to import all of these libraries for the notebook. If your code is ever having strange errors (when you didn't change it), make sure this block has been run so that your code is not trying to use libraries it doesn't yet know about.

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##Pie Chart

This is code for a simple pie chart. Try running it, and then changing some values to see what happens.

```
labels = 'Chocolate', 'Vanilla', 'Pistachio', 'Mango', 'Strawberry'
sizes = [12, 11, 4, 8, 7] # Notice these sizes do not add up to 100
explode = (0, 0, 0, 0.1, 0) # This array determines which slice is offset from
                            # the rest of the chart and how far it moves.

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90)
ax1.axis('equal')
 
ax1.set(title="Ice Cream Preferences")

plt.show()
```

##Bar Chart

This is a simple bar chart. What happens when you change the range of the y-axis?

```
N = 4
costs = (10, 12.4, 15, 31)
ind = np.arange(N)
width = 0.5

plt.bar(ind, costs, width)

plt.ylabel('Dollars')
plt.title('Meal Cost')
plt.xticks(ind, ('Noodles', 'Stew', 'Casserole', 'Steak'))
plt.yticks(np.arange(0, 45, 5)) # The first number is the lower bound of the y
                                # axis, second is the upper bound, third is the
                                # distance between tick marks.
plt.xticks(rotation=90) #rotates text for x-axis labels

plt.show()
```

##Line Graph

Here is a line graph. See what happens if you reorder the points so that the x-axis is not in ascending order.

```
# When you have two lists like this, the first element of one goes with the
# first element of the other, and so on.

plt.plot(
  [1, 2, 3, 4, 5, 6, 8, 9, 10], # These are the x values
  [8, 5, 6, 5.5, 4, 3, 2.5, 2, 4] # These are the y values
)
plt.axis([0, 11, 0, 9])
plt.show()
```

##Scatter Plot

This is one of the simpler plots to construct. In many cases, a scatter plot is the first visualization you should try when exploring a dataset.

```
plt.scatter(
  [1, 1, 2, 2, 3, 3, 4, 5, 6, 6, 6, 7, 8, 9, 10,10],
  [7, 7.5, 7.5, 5.5, 6, 5, 4.5, 5, 4, 3, 3.5, 3.5, 2, 3, 2.5,1]
)
plt.axis([0, 11, 0, 9])
plt.show()
```

##Heatmap

In this case, we are loading data in from a separate source. The amount of data represented in a heatmap is rarely small enough to list by hand, so we import it from separate files to keep our code short.

```
# You can also load the data as lists, just like the scatterplot.
# For most heatmaps, this will create excessively long code, so 
# It is better to import a dataset from another file.

iris = sns.load_dataset("iris")

sns.kdeplot(
    iris.sepal_width, # This just is the name of a standard python list, like
                      # the ones in the scatterplot above.
    iris.sepal_length,
    cmap="Reds",
    shade=True,
    shade_lowest=False
)
plt.show()
```
