##Exercise 1: Desserts

What is the second least popular dessert?

```
#A sweet shop serves 13 types of chocolate desserts. Here are the selling
# numbers for one evening:

#Lava Cake: 14
#Mousse: 5
#Chocolate Cake: 12
#Ice Cream: 19
#Truffles: 6
#Brownie: 8
#Chocolate Chip Cookie: 12
#Chocolate Pudding: 9
#Souffle: 10
#Chocolate Cheesecake: 17
#Chocolate Chips: 2
#Fudge: 9
#Mochi: 13
```

###Solution

One of the simpler ways to compare numerical data is with a bar chart, since you can check visually how tall each bar is compared to the others.

```
N = 13
num_sold = (14, 5, 12, 19, 6, 8, 12, 9, 10, 17, 2, 9, 13)
ind = np.arange(N)
width = 0.5

plt.bar(ind, num_sold, width)

plt.ylabel('Number Sold')
plt.title('Desserts sold in one evening')
plt.xticks(ind, ('Lava Cake', 
                 'Mousse', 
                 'Chocolate Cake', 
                 'Ice Cream', 
                 'Truffles', 
                 'Brownie', 
                 'Chocolate Chip Cookie', 
                 'Chocolate Pudding', 
                 'Souffle', 
                 'Chocolate Cheesecake', 
                 'Chocolate Chips', 
                 'Fudge', 
                 'Mochi'))
plt.yticks(np.arange(0, 20, 5))
plt.xticks(rotation=90)

plt.show()
```

The second least popular dessert is Mousse.

##Exercise 3: Frogs

##Exercise 4: Halloween Candy

Drawing at random, what is the percent chance that you will get a Snickers?

```
#A bag of halloween candy has these amounts of the various chocolate bars:

#Kit Kat: 52
#Snickers: 39
#Milky Way: 78
#Toblerone: 13
#Twix: 78
```

###Solution

Since we want to find percentages, we can use a pie chart here to make Matplotlib do the calculation for us.

```
labels = 'Kit Kat', 'Snickers', 'Milky Way', 'Toblerone', 'Twix'
sizes = [52, 39, 78, 13, 78] # Notice these sizes do not add up to 100
explode = (0, 0.1, 0, 0, 0) # This array determines which slice is offset from
                            # the rest of the chart and how far it moves.

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90)
ax1.axis('equal')

ax1.set(title="Halloween Candy")

plt.show()
```

There is a 15 percent chance you will get a Snickers.

##Exercise 5: Flights

Is there more variety in plane size/fullness now, or in 1952?

```
flights = sns.load_dataset("flights")
#flights.year gives a list of the years of various flights
#flights.passengers gives the number of passengers per flight
```

###Solution

A heatmap shows the variety in a very intuitive visual way. This could also be solved with a scatterplot, however.

```
sns.kdeplot(
    flights.year,
    flights.passengers,
    cmap="Blues",
    shade=True,
    shade_lowest=False
)
plt.show()
```

There is more variety now.
